```python
import pandas as pd
import numpy as np
from functools import reduce
```

```python
infile = r'files/movie_metadata.csv'
```

```python
#1.1. Download the data set movie_metadata.csv, which contains data about films from IMDb
#(Internet Movie Database).

def csv_to_df(infile, headers = []):
    if len(headers) == 0:
        df = pd.read_csv(infile)
    else:
        df = pd.read_csv(infile, header = None)
        df.columns = headers

    for r in range(10):
        try:
            df.rename( columns={'Unnamed: {0}'.format(r):'Unnamed{0}'.format(r)}, inplace=True )
        except:
            pass
    return df

df = csv_to_df(infile).rename_axis('id')
```

```python
#1.2. The duration column contains data on the film length. How many missing values are there
#in this column?
nan_duration = df['duration'].isna().sum()

nan_duration
```

```python
#1.3. Replace the missing values in the duration column with the median value for this column.
### Replacing NaN values on duration column ###

#Get avg from duration
duration_avg = df['duration'].mean()

# Setting duration_avg on NaN duration rows
df.duration = df.duration.fillna(duration_avg)
```

```python
# 1.4. What is the average film length? Give the answer as a floating-point figure rounded to
# two decimal places.
# Rounded duration avg
df['duration'] = df['duration'].round(2)

df['duration']
```

```python
### Creating a new column named: movie_duration_category ###
# 1.5. Create a movie_duration_category column, which will contain three categories
# depending on the film length:
result = []
for value in df['duration']:
    if value < 90:
        result.append(1)
    elif value >= 90 and value <= 120:
        result.append(2)
    else:
        result.append(3)

df['movie_duration_category'] = result
df['movie_duration_category']
```

```python
### New table ###
# 1.6. Build a summary table for films released after 2000 (inclusively), to list the numbers of
# films
newdf = df[(df.title_year >= 2000)]
newdf['title_year'] = pd.to_datetime(newdf['title_year'], format='%Y').dt.strftime('%Y')
```

```python
# 1.7. How many films between 90 minutes and two hours long were released in 2008?
over_2008 = newdf[(newdf.title_year == '2008') & (newdf.movie_duration_category == 2)]
total_over_2008 = len(over_2008)
total_over_2008
```

```python
# 1.8. The plot_keywords column holds keywords characterizing the film's plot. Using the data
# in this column, create a column called movie_plot_category, to contain four categories
# depending on the key words in the column:
df['plot_keywords'] = df['plot_keywords'].fillna('no_keywords')

key_words = ['love', 'death']

def common_member(a, b):
    key_words_set = set(a)
    plot_keywords_set = set(b)
 
    if (key_words_set & plot_keywords_set):
        commom = key_words_set & plot_keywords_set
        if len(commom) > 1:
            new_category = 'love_and_death'
        else :
            new_category = reduce(lambda a, b : a+', '+b , commom)
    else:
        new_category = 'other'
    
    return new_category

plot_cat = []
for value in df['plot_keywords']:
    plot_keywords = value.split('|')
    plot_cat.append(common_member(key_words, plot_keywords))

df['movie_plot_category'] = plot_cat
```

```python
# 1.9. The imdb_score column shows a viewer rating for the film. Build a table to reflect the
# average rating of films depending on which movie_plot_category category they belong to.
imdb_score_avg = df.groupby(['movie_plot_category'])['imdb_score'].mean()

imdb_score_avg
```

```python
# 1.10. What is the average rating of films in the "love" category? Give the answer as a floating point figure rounded to two decimal places.
love_imdb_score_avg = imdb_score_avg['love'].round(2)

love_imdb_score_avg
```

```python
# 1.11. The budget column contains the film's budget. What is the median budget for all the films listed? Give the answer as an integer.
df['budget'] = df['budget'].str.replace('$', '')
df['budget'] = df['budget'].astype(float)
df['budget'] = df['budget'].astype(np.int64)

budget_avg = df['budget'].mean().round(0).astype(int)

budget_avg
```
